```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
page_w = 1200
page_h = 1000
r = np.zeros((page_h, page_w), dtype=np.uint8)
ox = page_w // 2
oy = 0

for x in range(page_h):
    for y in range(page_w):
        if 0 < (x)**2 + (y-ox)**2 < 4000:
            r[x, y] = 255
            # r[150:300, 150:300] = 255
plt.imshow(r, cmap="gray")
plt.show()
```

```python
g = np.zeros((page_h, page_w), dtype=np.uint8)
for x in range(page_h):
    for y in range(page_w):
        if 0 < (x)**2 + (y-ox)**2 < 2000:
            g[x, y] = 255
            # r[150:300, 150:300] = 255
plt.imshow(r, cmap="gray")
plt.show()
```

```python
b = np.zeros((page_h, page_w), dtype=np.uint8)
for x in range(page_h):
    for y in range(page_w):
        if 0 < (x)**2 + (y-ox)**2 < 4000:
            b[x, y] = 255
            # r[150:300, 150:300] = 255
plt.imshow(r, cmap="gray")
plt.show()
```

```python
rainbow = cv2.merge([r, g, b])
plt.imshow(rainbow)
plt.show()
```

```python
b = np.zeros((300, 300), dtype=np.uint8)
b[150:300, 0:150] = 255
plt.imshow(b, cmap="gray")
plt.show()
```

```python
logo = cv2.merge([r, g, b])
plt.imshow(logo)
plt.show()
```
