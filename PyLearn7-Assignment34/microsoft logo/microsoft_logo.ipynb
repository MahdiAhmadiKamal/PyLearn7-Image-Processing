```python
import cv2
import  as np
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
```

```python
page_w = 2008
page_h = 900
r = np.zeros((page_h, page_w), dtype=np.uint8)
g = np.zeros((page_h, page_w), dtype=np.uint8)
b = np.zeros((page_h, page_w), dtype=np.uint8)
```

```python
for x in range(page_h):
    for y in range(page_w):
        r[x, y] = 87
        g[x, y] = 83
        b[x, y] = 82

img = cv2.merge([r, g, b])
plt.imshow(img, cmap="gray")
plt.show()
```

```python
for x in range(page_h):
    for y in range(page_w):
        if 299 < x < 443 and 299 < y < 443:
            r[x, y] = 242
            g[x, y] = 80
            b[x, y] = 34
        if 457 < x < 600 and 299 < y < 443:
            r[x, y] = 0
            g[x, y] = 164
            b[x, y] = 239
        if 299 < x < 443 and 457 < y < 600:
            r[x, y] = 127
            g[x, y] = 186
            b[x, y] = 0
        if 457 < x < 600 and 457 < y < 600:
            r[x, y] = 255
            g[x, y] = 185
            b[x, y] = 0
        

img = cv2.merge([r, g, b])


plt.imshow(img, cmap="gray")
plt.show()
```

```python

img = Image.fromarray(img)

draw = ImageDraw.Draw(img)
font = ImageFont.truetype("input/segoeuithis.ttf", 238)

draw.text((670, 280), "Microsoft", fill=(255, 255, 255) , font=font, stroke_width=3)
img = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

cv2.imwrite("output/result.png", img)
```
