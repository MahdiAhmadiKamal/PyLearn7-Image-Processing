```python
from math import sqrt
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
page_w = 2008
page_h = 900
r = np.zeros((page_h, page_w), dtype=np.uint8)
g = np.zeros((page_h, page_w), dtype=np.uint8)
b = np.zeros((page_h, page_w), dtype=np.uint8)
```

```python
for x in range(page_h):
    for y in range(page_w):
        r[x, y] = 87
        g[x, y] = 83
        b[x, y] = 82

img = cv2.merge([r, g, b])
plt.imshow(img, cmap="gray")
plt.show()
```

```python
for x in range(page_h):
    for y in range(page_w):
        if 301 < x < 445 and 299 < y < 443:
            r[x, y] = 242
            g[x, y] = 80
            b[x, y] = 34
        if 457 < x < 600 and 299 < y < 443:
            r[x, y] = 0
            g[x, y] = 164
            b[x, y] = 239
        if 299 < x < 443 and 456 < y < 599:
            r[x, y] = 127
            g[x, y] = 186
            b[x, y] = 0
        if 457 < x < 600 and 456 < y < 599:
            r[x, y] = 255
            g[x, y] = 185
            b[x, y] = 0
        

img = cv2.merge([r, g, b])
plt.imshow(img, cmap="gray")
plt.show()
```

```python

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2.imwrite("output/microsoft_logo.jpg", img)
```
