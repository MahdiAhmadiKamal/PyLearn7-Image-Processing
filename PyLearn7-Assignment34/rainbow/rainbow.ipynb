```python
from math import sqrt
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
page_w = 1200
page_h = 600
r = np.zeros((page_h, page_w), dtype=np.uint8)
ox = page_w // 2
oy = 0

for x in range(page_h):
    for y in range(page_w):
        if 0 < sqrt((x)**2 + (y-ox)**2) < 48 or 124 < sqrt((x)**2 + (y-ox)**2) < 196:
            r[x, y] = 255
            
plt.imshow(r, cmap="gray")
plt.show()
```

```python
g = np.zeros((page_h, page_w), dtype=np.uint8)
for x in range(page_h):
    for y in range(page_w):
        if 0 < sqrt((x)**2 + (y-ox)**2) < 24 or 72 < sqrt((x)**2 + (y-ox)**2) < 148:
            g[x, y] = 255
        if 148 < sqrt((x)**2 + (y-ox)**2) < 172:
            g[x, y] = 165
            
plt.imshow(r, cmap="gray")
plt.show()
```

```python
b = np.zeros((page_h, page_w), dtype=np.uint8)
for x in range(page_h):
    for y in range(page_w):
        if 0 < sqrt((x)**2 + (y-ox)**2) < 96:
            b[x, y] = 255
            
plt.imshow(r, cmap="gray")
plt.show()
```

```python
rainbow = cv2.merge([r, g, b])
plt.imshow(rainbow)
plt.show()
cv2.imwrite("output/r.jpg", rainbow)
```
