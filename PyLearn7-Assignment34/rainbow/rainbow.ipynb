```python
from math import sqrt
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
page_w = 900
page_h = 600
r = np.zeros((page_h, page_w), dtype=np.uint8)
ox = page_w // 2
oy = 0

for x in range(page_h):
    for y in range(page_w):
        if 0 < sqrt((x)**2 + (y-ox)**2) < 96 or 248 < sqrt((x)**2 + (y-ox)**2) < 392:
            r[x, y] = 255
        if 392 < sqrt((x)**2 + (y-ox)**2):
            r[x, y] = 255
            
plt.imshow(r, cmap="gray")
plt.show()
```

```python
g = np.zeros((page_h, page_w), dtype=np.uint8)
for x in range(page_h):
    for y in range(page_w):
        if 0 < sqrt((x)**2 + (y-ox)**2) < 48 or 144 < sqrt((x)**2 + (y-ox)**2) < 296:
            g[x, y] = 255
        if 296 < sqrt((x)**2 + (y-ox)**2) < 344:
            g[x, y] = 165
        if 392 < sqrt((x)**2 + (y-ox)**2):
            g[x, y] = 255
            
plt.imshow(r, cmap="gray")
plt.show()
```

```python
b = np.zeros((page_h, page_w), dtype=np.uint8)
for x in range(page_h):
    for y in range(page_w):
        if 0 < sqrt((x)**2 + (y-ox)**2) < 192:
            b[x, y] = 255
        if 392 < sqrt((x)**2 + (y-ox)**2):
            b[x, y] = 255   

plt.imshow(r, cmap="gray")
plt.show()
```

```python
rainbow = cv2.merge([r, g, b])
plt.imshow(rainbow)
plt.show()
rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
rainbow = cv2.rotate(rainbow, rotateCode=1)
cv2.imwrite("output/rainbow.jpg", rainbow)
```
